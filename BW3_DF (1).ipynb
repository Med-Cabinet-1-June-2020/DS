```
import pandas as pd 
import numpy as np
import urllib.request, json
import requests
```

```
url = ('http://strainapi.evanbusse.com/OEdbKsl/strains/search/all')
```

```
import requests
response = requests.get(url)
data = response.json()
```

```
data
```

```
df = pd.DataFrame(data)

df.head()
```

```
df.columns
```

```
df = df.T

df.head()
```

```
df = df.reset_index().set_index('id')

df.head()
```

```
df = df.rename(columns={"index": "Strain"})
df = df.reset_index()
df.head()
```

```
df.shape
```

```
df2 = pd.json_normalize(df['effects'])
df2 = df2.reset_index()
df2.head()
```

```
df2.shape
```

```
df3 = pd.concat([df,df2], axis=1)

df3.head(10)
```

```
df3 = df3.drop(columns=['effects','index']).set_index('id')

df3.shape
df3.head()
```

```
df3.shape
```

```
df4 = pd.read_csv('https://raw.githubusercontent.com/Med-Cabinet-1-June-2020/DS/master/cannabis.csv')

df4.head()
```

```
df4['Strain'] = df4['Strain'].str.replace('-', ' ')

df4.head()
```

```
df4.shape
```

```
df5 = pd.merge(df3, df4, how='inner', on='Strain')

df5.head()
```

```
df6 = df5.drop(columns=['Effects', 'Flavor', 'race'])
df6 = df6.rename(columns={"flavors": "Flavors", "positive": "Positive",
                          "negative": "Negative", "medical": "Medical"})
df6.head()
```

```
df6.shape
```

```
df6.to_csv('final_cannabis.csv', index=False) 
```
